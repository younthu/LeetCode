```python
#https://leetcode.com/problems/4sum/
from typing import List
from bisect import bisect_left, bisect_right
import pysnooper
from math import ceil

class Solution:
    """
    三层二分查找
    """
    @pysnooper.snoop()
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        l = len(nums)
        last_num = nums[l - 1]
        predict_a = bisect_right(nums, target / 4.0)
        index_a = 0
        result = []
        a = 0
        b = 0
        c = 0
        d = 0
        
        while index_a < predict_a and nums[index_a] * 4 <= target and nums[index_a] + last_num*3 >=target:
            a = nums[index_a]
            predict_b = bisect_right(nums, target - a/3.0)
            
                
            index_b = index_a + 1
            
            
            while index_b < predict_b and b * 3 <= target - a and b + last_num*2 >= target - a:
                b = nums[index_b]
                predict_c = bisect_right(nums, (target - a - b)/2.0)
                index_c = index_b + 1
                
                
                while index_c < predict_c:
                    c = nums[index_c]
                    predict_d = bisect_left(nums,target - a - b - c)
                    
                    index_d = index_c + 1
                    
                    while index_d < predict_d:
                        d = nums[index_d]
                        
                        if a + b + c + d > target:
                            break
                        elif a + b + c + d == target:
                            result.append([a,b,c,d])
                            break
                        index_d += 1
                        pass
                    index_c += 1
                    pass
                index_b += 1
                pass
            index_a +=1
        
        return result
                            
                
        
        
                        
```

```python
s = Solution()
s.fourSum([1, 0, -1, 0, -2, 2], 0) 
```
