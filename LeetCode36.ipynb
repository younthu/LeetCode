```python
from typing import List
from collections import Counter
import pysnooper

class Solution:
    @pysnooper.snoop()
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        for row in range(9):
            cn = Counter()
            cn2 = Counter()
            for col in range(9):
                c1 = board[row][col]
                c2 = board[col][row]
                if c1 == ".":
                    pass
                else:
                    if cn[c1] > 0:
                        return False
                    else:
                        cn[c1]=1
                        
                if c2 == ".":
                    pass
                else:
                    if cn2[c2] > 0:
                        return False
                    else:
                        cn2[c2]=1
            
        
        for row in range(3):
            for col in range(3):
                cn = Counter()
                
                for i in range(3):
                    for j in range(3):
                        c = board[row*3 +i][col*3 +j]
                        if c == ".":
                            pass
                        else:
                            if cn[c] >0:
                                return False
                            else:
                                cn[c]=1
                
                    
        return True
        
    
```
