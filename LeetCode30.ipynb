```python
from collections import Counter
from typing import List
import pysnooper

class Solution:
    @pysnooper.snoop()
    def findSubstring(self, s: str, words: List[str]) -> List[int]:
        cn = Counter("".join(words))
        if len(s)==0 or len(words)==0:
            return []
        
        l = len(words) * len(words[0])
        wsize = len(words)
        
        wl = len(words[0])
        ret = []
        
        if l > len(s):
            return []
        
        for idx in range(l):
            c = s[idx]
            cn[c]-=1
        
        words.sort()
        for idx in range(len(s) - l + 1):
            if not any(cn.values()): # likely found the match
                temp = [s[index*wl + idx:idx + index*wl+wl] for index in range(wsize)]
                temp.sort()
                if temp == words:
                    ret.append(idx)
            else:
                pass
            
            if idx != len(s) - l:
                cn[s[idx]] += 1
                cn[s[idx+l]]-= 1
                
        return ret
```

```python
s = Solution()
s.findSubstring("foobarfoobarf",['bar','foo'])
```

```python
s = Solution()
s.findSubstring("barfoofoobarthefoobarman",["bar","foo","the"])
```
