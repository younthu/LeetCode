```python
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        self.s2(self, nums, target)
    def s2(self, nums: List[int], target: int) -> List[int]:
        i = 0;
        a_len = len(nums)
        h = {k: v for v, k in enumerate(nums)}
        while(i < a_len):
            sec_num = target - nums[i]
            if sec_num in h and h[sec_num] != i:
                return [i, h[sec_num]]
            i+=1
    def s1(self, nums: List[int], target: int) -> List[int]:
        i = 0;
        while(i < len(nums)):
            sec_num = target - nums[i]
            j = i + 1
            while(j < len(nums)):
                if nums[j] == sec_num:
                    return [i,j]
                j+=1
            i+=1
```
