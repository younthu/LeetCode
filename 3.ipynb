```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        sub = []
        max_len = 0
        for c in s:
            if c in sub:
                idx = sub.index(c)
                sub = sub[idx+1:]
            sub.append(c)
            l = len(sub)
            max_len = l if l > max_len else max_len
        return max_len
        
```
