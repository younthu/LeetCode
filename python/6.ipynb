```python
class Solution:
    def convert(self, s: str, numRows: int) -> str:
        if numRows == 1:
            return s
        
        l = len(s)
        ret = ["a"] * l
        row = 0 # 当前行号
        col = 0 # col
        cur = 0 # s的游标
        count = 0 # 已经排好多少字
        
        # 第0行
        s_idx = 0 # s的游标,每次找到行首游标开始跳
        cur = 0 #ret的游标, 一个挨一个填充
        while s_idx < l:
            ret[cur] = s[s_idx]
            cur += 1
            s_idx += numRows * 2 -2
            count += 1
        
        # 第1至numRows - 2行
        row = 1
        while row < numRows - 1:
            cur = count
            s_idx = row # 起始点和row一致
            while s_idx < l:
                #填充左边的一个
                ret[cur] = s[s_idx]
                count += 1
                
                # 填充右边的那个
                s_idx2 = s_idx + (numRows - row -1) * 2
                if s_idx2 < l:
                    ret[cur + 1] = s[s_idx2]
                    count +=1
                    
                cur += 2
                s_idx += numRows * 2 -2
            row +=1
        
        # 最后一行
        cur = count # ret游标
        s_idx = numRows - 1 # s游标
        while s_idx < l:
            ret[cur] = s[s_idx]
            cur+=1
            count+=1
            s_idx += numRows * 2 -2
        
        return "".join(ret)
            
                
        
        
        
        
         
```
