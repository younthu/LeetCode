```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        l = 0
        for c in s:
            if l > 0 and self.match(stack[l-1], c):
                stack.pop()
                l -= 1
            else:
                stack.append(c)
                l += 1
        return l == 0
    def match(self, c1, c2):
        return (c1 == "(" and c2 == ")") or (c1 == "{" and c2 == "}") or (c1 == "[" and c2 == "]")
            
```
