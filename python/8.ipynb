```python
class Solution:
    def myAtoi(self, s: str) -> int:
        prefix = 1
        ret = 0
        if not s:
            return 0

        # 去除前导空格
        while s and s[0] == ' ':
            s = s[1:]
        
        # 前缀
        if s and s[0] == '-':
            prefix = -1
            s = s[1:]
        elif s and s[0] == '+':
            s = s[1:]
        # 去除前导0
        while s and s[0] == '0':
            s = s[1:]
            
        # 遍历所有数字
        l = len(s)
        idx = 0
        while idx < l:
            if s[idx] in "0123456789":
                ret *= 10
                ret += int(s[idx])
                idx += 1
            else:
                break
        ret *= prefix
        if ret > (1<<31) -1:
            ret = (1<<31) -1
        if ret < -1<<31:
            ret = -1<<31
        return ret
```
