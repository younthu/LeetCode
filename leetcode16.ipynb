```python
# https://leetcode.com/problems/3sum-closest/
import pysnooper
from typing import List
from bisect import bisect_left
class Solution:
    """
    思路: 数组排序，然后三个游标, 一个从左往右，一个从右往左(循环反复)，一个通过二分定位
    """
    @pysnooper.snoop()
    def threeSumClosest(self, nums: List[int], target: int) -> int:
        l = len(nums)
        left = 0
        right = l - 1
        nums.sort()
        minimum_diff = -1
        result = 0
        
        while(left < l - 1):
            right = l - 1

            while(right > left + 1):
                t = - nums[left] - nums[right] + target   
                index1 = bisect_left(nums, t, left + 1, right)
                if index1 == right:
                    s = nums[left] + nums[right] + nums[right - 1]
                    if minimum_diff == -1:
                        minimum_diff = s - target if s >target else target - s
                        result = s
                    else:
                        diff = s-target if s>target else target-s
                        if diff < minimum_diff:
                            minimum_diff = diff
                            result = s
                elif index1 == left + 1:
                    s = nums[left] + nums[right] + nums[left + 1]
                    if minimum_diff == -1:
                        minimum_diff = s - target if s >target else target - s
                        result = s
                    else:
                        diff = s-target if s>target else target-s
                        if diff < minimum_diff:
                            minimum_diff = diff
                            result = s
                else:
                    s = nums[left] + nums[right] + nums[index1 -1]
                    s1 = nums[left] + nums[right] + nums[index1]
                    s2 = nums[left] + nums[right] + nums[index1 + 1]
                    diff = 0
                    ts = s
                    if target < s:
                        diff = s - target
                        ts = s
                    elif s <= target < s1:
                        if target -s < s1 - target:
                            diff = target - s 
                            ts = s
                        else:
                            diff = s1 - target
                            ts = s1
                    else:
                        diff = target - s1
                        ts = s1

                    if minimum_diff == -1:
                        minimum_diff = s - target if s >target else target - s
                        result = ts
                    else:
                        if diff < minimum_diff:
                            minimum_diff = diff
                            result = ts
                right -= 1
            left+=1
        
        
        return result        
            
        
```
